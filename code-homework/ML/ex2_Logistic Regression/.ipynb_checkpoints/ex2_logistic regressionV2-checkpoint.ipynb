## 读取数据并可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
data = pd.read_csv('ex2data1.txt',names = ['exam 1 score','exam 2 score','admitted?'])
data.head()
```

```python
df = data.copy()
col = df.shape[1]
x = df.iloc[:,0:col - 1]
y = df.iloc[:,col - 1 :col]
X = x.copy()
X.insert(0,'one',1)
x = np.array(x)
y = np.array(y)
X = np.array(X)
x_1 = x[:,0]
x_2 = x[:,1]
x,y,X,x_1,x_2,x_1,y.shape
```

```python
"""将原数据分组"""
pos = []
neg = []
for i in range(x.shape[0]):#选出正类和负类的索引
    if y[i] == 1:
        pos.append(i)
    else:
        neg.append(i)
pos_x_1 = x_1[pos]#一维矩阵不能用于scatter中，要转换为一维数组或列表
pos_x_2 = x_2[pos]
neg_x_1 = x_1[neg]
neg_x_2 = x_2[neg]
```

```python
fig,ax = plt.subplots()
ax.scatter(pos_x_1,pos_x_2,label = 'admitted',marker = 'o')
ax.scatter(neg_x_1,neg_x_2,label = 'not admitted',marker= 'x')
ax.legend(loc = 1)
```

## sigmoid函数

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
n = np.arange(-10.,10.,0.1)
sigmoid(n)
fig = plt.figure()
plt.plot(n,sigmoid(n))
plt.show()
```

## 代价函数

```python
theta = np.zeros(3)
def costfunction(theta, X,y):
    theta = theta.reshape(X.shape[1],1)
    cost_pos = -y * np.log(sigmoid(X @ theta))
    cost_neg = -(1-y) * np.log(1 - sigmoid(X @ theta))
    cost = 1/len(X)*(cost_pos + cost_neg)
    return np.sum(cost)
costfunction(theta,X,y)
```

```python
y.shape,X.shape,theta.shape
```

## 梯度下降

```python
"""一个梯度步长"""
def gradient(theta,X,y):
    theta = theta.reshape(X.shape[1],1)
    h = X @ theta
    erro = sigmoid(h) - y
    grad = (erro.T @ X)/len(X)
    return grad.ravel()
```

```python
gradient(theta,X,y)
```

```python
"""迭代10000次"""
costs = [costfunction(theta,X,y)]
for i in range(100000):
    theta = theta.reshape(X.shape[1],1)
    temp = theta - 0.01*gradient(theta, X,y).reshape(X.shape[1],1)
    theta = temp
    costs.append(costfunction(theta,X,y))
costs
```

```python
fig  =plt.figure()
plt.plot(costs)
theta
##在alpha=0.01 迭代10000次后，代价来回跳动，依然很难收敛
```

```python
import scipy.optimize as opt
result = opt.minimize(fun=costfunction, x0=theta, args=(X,y), method='TNC')
result  ##参数类型难搞
theta = result.x[:,np.newaxis]
xx = np.arange(20,100,0.1)
theta = theta / theta[2]
yy = -theta[0] - theta[1] *xx
fig = plt.figure()
plt.plot(xx,yy)
plt.scatter(pos_x_1,pos_x_2,label = 'admitted',marker = 'o')
plt.scatter(neg_x_1,neg_x_2,label = 'not admitted',marker= 'x')
plt.show()
```

```python
"""用sklearn中的逻辑回归"""
```

```python
import sklearn.linear_model as lm
classfiser =lm.LogisticRegression(solver='liblinear', C=100)
y = y.ravel()
classfiser.fit(x,y)
x_min,x_max = min(X[:,1]) - 1.0, max(X[:,1]) + 1.0
y_min,y_max = min(X[:,2]) - 1.0, max(X[:,2]) + 1.0
x_values,y_values = np.meshgrid(np.arange(x_min,x_max,0.01),np.arange(y_min,y_max,0.1))
mesh_output = classfiser.predict(np.c_[x_values.ravel(),y_values.ravel()])
mesh_output = mesh_output.reshape(x_values.shape)
plt.figure()
plt.pcolormesh(x_values,y_values,mesh_output,cmap=plt.cm.gray)
plt.scatter(pos_x_1,pos_x_2,label = 'admitted',marker = 'o')
plt.scatter(neg_x_1,neg_x_2,label = 'not admitted',marker= 'x')
plt.show()
```

# 二、逻辑回归正则化

```python
data2 = pd.read_csv("ex2data2.txt",names = ["text1","text2","quality"])
data2.head()
```

```python
pos = data2[data2['quality'].isin([1])]
neg = data2[data2['quality'].isin([0])]

fig, ax = plt.subplots()
ax.scatter(pos['text1'], pos['text2'], s=50, c='b', marker='o', label='y = 1')
ax.scatter(neg['text1'], neg['text2'], s=50, c='r', marker='x', label='y = 0')
ax.legend()
ax.set_xlabel('text1')
ax.set_ylabel('text2')
plt.show()
```

## 特征映射

```python
df2 = data2.copy()
x1 = df2["text1"]
x2 = df2["text2"]
x = np.array(df2.iloc[:,0:2])
def feature_map(x1,x2,power): 
    for i in np.arange(power+1):
        for j in np.arange(power+1):
            df2['F' + str(i) + str(j)] = np.power(x1, i) * np.power(x2, j)
    col = df2.shape[1]
    X = df2.iloc[:,3:col]
    X.insert(0,'one',1)
    X = np.array(X)
    y = np.array(df2.iloc[:,2:3])
    return df2,X,y
df2,X,y = feature_map(x1,x2,power = 5)
df2.head(),X,y
```

## 正则化代价函数

```python
theta = np.zeros(X.shape[1])
X.shape,y.shape,theta.shape
```

```python
def costfunctionreg(theta, X,y,c=1):
    theta = theta.reshape(X.shape[1],1)
    reg = c/len(X)*np.sum(theta[1:,:]**2)
    return costfunction(theta,X,y)+reg
costfunction(theta,X,y)
```

## 正则化剃度下降

```python
"""正则化梯度步长"""
def gradientreg(theta,X,y,c=1):
    theta = theta.reshape((X.shape[1],1))
    reg = 1/len(X)*theta[1:,:]
    reg = np.vstack([np.zeros((1,1)),reg])
    return ((gradient(theta, X,y).reshape(X.shape[1],1) + reg)).ravel()
gradientreg(theta,X,y,c=1).shape
```

```python
"""迭代11000次"""
def gradientreg_decent(theta,X,y,itera,c=1):
    costs = [costfunctionreg(theta,X,y,c=1)]
    for i in range(itera):
        theta = theta.reshape(X.shape[1],1)
        theta = theta - 0.005*gradientreg(theta,X,y,c=1).reshape(X.shape[1],1)
        costs.append(costfunctionreg(theta,X,y,c=1))
    return theta,costs
theta,costs = gradientreg_decent(theta, X, y,  itera = 11000, c=1)
plt.figure()
plt.plot(costs)
plt.show()
```

```python
theta,costs[-1]
```

## 拟合参数

```python
def opt_min(X,y):
    init_theta = np.zeros(X.shape[1])
    res = opt.minimize(fun=costfunctionreg,
                       x0=init_theta,
                       args=(X, y, 0.1),
                       method='TNC',
                       jac=gradientreg,
                       options={'maxiter': 10000})
    theta = res.x
    h = sigmoid(X @ theta[np.newaxis,:].T)
    h[h>=0.5] = 1
    h[h<0.5] = 0
    from sklearn.metrics import classification_report#这个包是评价报告
    return res,print(classification_report(y.ravel(), h.ravel()))
```

```python
opt_min(X,y)
```

```python
"""逻辑回归，并返回准确率报告"""
def logistic_regression(X, y, itera):
    theta = np.zeros((X.shape[1], 1))
    final_theta,_ = gradientreg_decent(theta,X, y, itera, c=1 )
    classifier = sigmoid(X @ final_theta)#这就是训练好的分类器
    y_pred = np.zeros((X.shape[0], 1))
    y_pred[classifier >= 0.5] = 1
    y_pred[classifier <= 0.5] = 0 #二分类中置信度大于0.5的认为是正类，反之为负类
    from sklearn.metrics import classification_report#这个包是评价报告
    return print(classification_report(y, y_pred))#返回分类器的准确率   
```

# 随机参数
上面不管是用高级优化法优化的模型，还是迭代了40000次的梯度下降的模型，在1的准确率都不高。我怀疑是初始参数的问题，下面尝试用随机的参数试试

```python
def randinit_theta(n, epsilon = 0.12):
    return np.random.uniform(-epsilon, epsilon, n)
```

```python
df2,X,y = feature_map(x1,x2,10)
init_theta = randinit_theta(X.shape[1])
res = opt.minimize(fun=costfunctionreg,
                   x0=init_theta,
                   args=(X, y, 0.1),
                   method='TNC',
                   jac=gradientreg,
                   options={'maxiter': 10000})
theta = res.x
h = sigmoid(X @ theta[np.newaxis,:].T)
h[h>=0.5] = 1
h[h<0.5] = 0
from sklearn.metrics import classification_report#这个包是评价报告
res,print(classification_report(y.ravel(), h.ravel()))
```

```python
init_theta = randinit_theta(X.shape[1])
final_theta,_ = gradientreg_decent(theta,X, y, itera = 30000, c=1 )
classifier = sigmoid(X @ final_theta)#这就是训练好的分类器
y_pred = np.zeros((X.shape[0], 1))
y_pred[classifier >= 0.5] = 1
y_pred[classifier <= 0.5] = 0 #二分类中置信度大于0.5的认为是正类，反之为负类
print(classification_report(y, y_pred))#返回分类器的准确率
```
